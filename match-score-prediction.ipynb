### Importing Liabraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
columns = ['start_date', 'venue', 'innings', 'ball', 'batting_team', 'bowling_team', 'runs_off_bat', 'extras']
data = pd.read_csv('all_matches.csv', usecols=columns)
data.head(20)
```

```python
data['runs'] = data.iloc[:,6:].sum(axis=1)
data.head(20)
```

```python
data['batting_team'].unique()
```

```python
main_teams = ['Kolkata Knight Riders', 'Royal Challengers Bangalore',
       'Chennai Super Kings', 'Kings XI Punjab', 'Rajasthan Royals',
       'Delhi Daredevils','Mumbai Indians',
       'Sunrisers Hyderabad','Delhi Capitals']
```

```python
data = data[(data['batting_team'].isin(main_teams)) & (data['bowling_team'].isin(main_teams))]
```

```python
data['batting_team'].unique()
```

```python
data.head()
```

```python
# Deleting rows of Super Overs
```

```python
len(data[data['innings'] > 2])
```

```python
data = data.drop(data[data.innings > 2].index)
```

**As we can see that some grounds have different names, so we can replace it with a common name. For example:-**

**Wankhede Stadium, Mumbai** and **Wankhede Stadium** both are same so we will convert it to **Wankhede Stadium**.

```python
data["venue"]= data["venue"].str.replace("Wankhede Stadium, Mumbai", "Wankhede Stadium")
data["venue"]= data["venue"].str.replace("MA Chidambaram Stadium, Chepauk, Chennai", "MA Chidambaram Stadium")
data["venue"]= data["venue"].str.replace("Rajiv Gandhi International Stadium, Uppal", "Rajiv Gandhi International Stadium")
data["venue"]= data["venue"].str.replace("Punjab Cricket Association Stadium, Mohali", "Punjab Cricket Association Stadium")
data["venue"]= data["venue"].str.replace("Sardar Patel Stadium, Motera", "Sardar Patel Stadium")
data["venue"]= data["venue"].str.replace("Punjab Cricket Association IS Bindra Stadium, Mohali", "Punjab Cricket Association Stadium")
```

```python
data["venue"]= data["venue"].str.replace("MA Chidambaram Stadium, Chepauk", "MA Chidambaram Stadium")
```

```python
data['venue'].value_counts()
```

```python
data['venue'].value_counts().count()
```

```python
data['venue'] = data['venue'].str.lower()
```

```python
data.head(2)
```

**Null Values**

```python
data.isnull().sum()
```

```python
from datetime import datetime
data['start_date'] = data['start_date'].apply(lambda x: datetime.strptime(x, '%Y-%m-%d'))
```

```python
data.head(6)
```

```python
col = ['runs_off_bat', 'extras']
data = data.drop(col, axis=1)
data.head()
```

### Only keeping first 6 overs data

```python
data = data.drop(data[data['ball']>5.6].index)
data.head()
```

#### Encoding

```python
from sklearn.preprocessing import LabelEncoder
venue_encoder = LabelEncoder()
battingteam_encoder = LabelEncoder()
bowlingteam_encoder = LabelEncoder()
```

```python
data['venue'] = venue_encoder.fit_transform(data['venue'])
data['batting_team'] = battingteam_encoder.fit_transform(data['batting_team'])
data['bowling_team'] = bowlingteam_encoder.fit_transform(data['bowling_team'])
data.head()
```

```python
data = data.drop('start_date', axis=1)
data.head(1)
```
